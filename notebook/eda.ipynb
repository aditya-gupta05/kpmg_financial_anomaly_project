```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/creditcard_cleaned.csv")
```

```python
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- Basic Statistics ---")
print(df.describe())

print("\n--- Class Distribution ---")
print(df['Class'].value_counts())
print(df['Class'].value_counts(normalize=True))
```

```python
# Coorrelation Heatmap
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Correlation Heatmap of Features", fontsize=14)
plt.show()
```

```python
# Histograms of Amount and Time
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['norm_Amount'], bins=50, ax=ax[0], kde=True)
ax[0].set_title("Distribution of Normalized Amount")

sns.histplot(df['norm_Time'], bins=50, ax=ax[1], kde=True)
ax[1].set_title("Distribution of Normalized Time")

plt.tight_layout()
plt.show()

# sns.histplot(df1['norm_Amount'], bins=50, ax=ax[0], kde=True)
# ax[0].set_title("Distribution of Normalized Amount")

# sns.histplot(df1['norm_Time'], bins=50, ax=ax[1], kde=True)
# ax[1].set_title("Distribution of Normalized Time")

# plt.tight_layout()
# plt.show()
```

```python
# Countplot of Fraud vs Non-Fraud Count
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Non-Fraud Transactions")
plt.xticks([0, 1], ["Non-Fraud", "Fraud"])
plt.show()
```

```python
# Boxplot of Amount by Class
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class', y='norm_Amount', data=df)
plt.title("Transaction Amount by Class")
plt.xticks([0, 1], ["Non-Fraud", "Fraud"])
plt.show()
```

```python
# Correlation of Features with Class
corr_with_target = corr['Class'].sort_values(ascending=False)
print("\n--- Correlation of Features with Class ---")
print(corr_with_target)
```
